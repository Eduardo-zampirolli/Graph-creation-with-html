```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
df = pd.read_csv('est-dir - parliament_data_by_year.csv')

# Clean data - handle missing/empty Tipo values
df['Tipo'] = df['Tipo'].fillna('Unknown')
df['Tipo'] = df['Tipo'].str.strip()

# Standardize case for consistent grouping
df['Tipo'] = df['Tipo'].str.upper()
df['Source'] = df['Source'].str.title()  # Capitalize first letter

# Convert Year to datetime for better plotting
df['Year'] = pd.to_datetime(df['Year'], format='%Y').dt.year
```

```python
# Load the data
#df = pd.read_csv('est-dir - parliament_data_by_year.csv')
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='Year', y='Count', hue='Source', 
             estimator='sum', errorbar=None)
plt.title('Notas taquigráficas coletadas por ano da Câmara e do Senado')
plt.ylabel('Total')
plt.xlabel('Ano')
plt.grid(True)
plt.tight_layout()
plt.savefig('total_activity_by_year.png')
plt.show()
```

```python
# Get top 10 activity types
tipo_counts = df.groupby('Tipo')['Count'].sum().sort_values(ascending=False)


# Define threshold for considering a category as "small" (e.g., top 5)
threshold = 6  # Keep top 5 categories, group the rest as "Outros"
top_categories = tipo_counts.head(threshold).index

# Create a new column for grouped categories
df['Tipo_Grouped'] = df['Tipo'].where(df['Tipo'].isin(top_categories), 'OUTROS')

# Now plot with the grouped categories
plt.figure(figsize=(12, 6))
sns.barplot(data=df, 
            x='Tipo_Grouped', 
            y='Count', 
            hue='Source', 
            estimator='sum', 
            errorbar=None,
            order=list(top_categories) + ['OUTROS'])  # Ensure consistent order

plt.title(f'Tipos das notas taquigráficas coletadas')
plt.xticks(rotation=45)
plt.ylabel('Total')
plt.xlabel('Tipo')
plt.tight_layout()
plt.savefig('activity_types_with_outros.png')
plt.show()
```

```python
# Compare Comissão and Plenário activities
key_metrics = ['COMISSÃO', 'PLENÁRIO', 'CPI']

plt.figure(figsize=(14, 8))
for i, metric in enumerate(key_metrics, 1):
    plt.subplot(2, 2, i)
    metric_df = df[df['Tipo'] == metric]
    sns.barplot(data=metric_df, x='Year', y='Count', hue='Source')
    plt.title(metric.title())
    plt.xticks(rotation=45)
    plt.tight_layout()
plt.suptitle('Comparison of Key Activity Types', y=1.02)
plt.savefig('key_metrics_comparison.png')
plt.show()
```

```python
senado_df = df[df['Source'] == 'Senado']
heatmap_data = senado_df.pivot_table(index='Tipo', 
                                    columns='Year', 
                                    values='Count', 
                                    aggfunc='sum')

plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, annot=True, fmt='g', cmap='YlGnBu')
plt.title('Senado Activity Heatmap by Year and Type')
plt.tight_layout()
plt.savefig('senado_heatmap.png')
plt.show()
```
